```python
df = df.rename(columns={
    'year':'Year',
    'month':'Month',
    'carrier':'Carrier_Code',
    'carrier_name':'Airline',
    'airport':'Airport_Code',
    'airport_name':'Airport',
    'arr_flights':'Total_Flights',
    'arr_del15':'Delayed_Flights',
    'carrier_ct':'Carrier_Delays',
    'weather_ct':'Weather_Delays',
    'nas_ct':'NAS_Delays',
    'security_ct':'Security_Delays',
    'late_aircraft_ct':'Late_Aircraft_Delays',
    'arr_cancelled':'Cancelled_Flights',
    'arr_diverted':'Diverted_Flights',
    'arr_delay':'Total_Delay_Minutes',
    'carrier_delay':'Carrier_Delay_Minutes',
    'weather_delay':'Weather_Delay_Minutes',
    'nas_delay':'NAS_Delay_Minutes',
    'security_delay':'Security_Delay_Minutes',
    'late_aircraft_delay':'Late_Aircraft_Delay_Minutes'
})
```

```python
df.columns
```

On-Time Flights

```python
df['On_Time_Flights'] = df['Total_Flights'] - df['Delayed_Flights']
```

Delay Percentage

```python
df['Delay_Percentage'] = (
    df['Delayed_Flights'] / df['Total_Flights']
) * 100
```

Cancellation Percentage

```python
df['Cancellation_Percentage'] = (
    df['Cancelled_Flights'] / df['Total_Flights']
) * 100
```

Average Delay per Delayed Flight

```python
df['Average_Delay_Per_Flight'] = (
    df['Total_Delay_Minutes'] /
    df['Delayed_Flights']
)
```

On-Time Percentage

```python
df['On_Time_Percentage'] = (
    df['On_Time_Flights'] /
    df['Total_Flights']
) * 100
```

Delay Severity

```python
def delay_level(delay):

    if delay < 15:
        return "Low"

    elif delay < 30:
        return "Moderate"

    elif delay < 60:
        return "High"

    else:
        return "Critical"

df['Delay_Severity'] = df['Average_Delay_Per_Flight'].apply(delay_level)
```
