Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
```

Load cleaned dataset

```python
df = pd.read_csv("../dataset/airline_operations_cleaned.csv")
```

Which airline handled the highest number of flights?

```python
top_airlines = (
    df.groupby("Airline")["Total_Flights"]
      .sum()
      .sort_values(ascending=False)
)

top_airlines
```

```python
top_airlines.plot(kind="bar", figsize=(12,5))

plt.title("Total Flights by Airline")
plt.xlabel("Airline")
plt.ylabel("Flights")

plt.xticks(rotation=90)

plt.show()
```

Which airline has the highest delay percentage?

```python
delay = (
    df.groupby("Airline")["Delay_Percentage"]
      .mean()
      .sort_values(ascending=False)
)

delay
```

```python
delay.plot(kind="bar", figsize=(12,5))

plt.title("Average Delay Percentage by Airline")
plt.ylabel("Delay %")

plt.xticks(rotation=90)

plt.show()
```

Top 10 busiest airports

```python
top_airports = (
    df.groupby("Airport")["Total_Flights"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_airports
```

```python
top_airports.plot(kind="barh", figsize=(10,6))

plt.title("Top 10 Airports by Flights")

plt.show()
```

Delay Reasons

```python
reasons = df[
    [
        "Carrier_Delays",
        "Weather_Delays",
        "NAS_Delays",
        "Security_Delays",
        "Late_Aircraft_Delays"
    ]
].sum()

reasons
```

```python
reasons.plot(kind="pie", autopct="%1.1f%%")

plt.ylabel("")

plt.title("Delay Reasons")

plt.show()
```

Flights by Year

```python
year = (
    df.groupby("Year")["Total_Flights"]
      .sum()
)

year
```

```python
year.plot(kind="bar")

plt.title("Flights by Year")

plt.show()
```

Which airlines have the best On-Time Performance?

```python
on_time = (
    df.groupby("Airline")["On_Time_Percentage"]
      .mean()
      .sort_values(ascending=False)
)

on_time
```

```python
plt.figure(figsize=(12,6))

on_time.plot(kind="bar")

plt.title("Average On-Time Performance by Airline")
plt.ylabel("On-Time %")
plt.xticks(rotation=90)

plt.show()
```

Top 10 Airports with Highest Delays

```python
airport_delay = (
    df.groupby("Airport")["Total_Delay_Minutes"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

airport_delay
```

```python
plt.figure(figsize=(10,6))

airport_delay.plot(kind="barh")

plt.title("Top 10 Airports by Delay Minutes")

plt.show()
```

Cancellation Analysis

```python
cancel = (
    df.groupby("Airline")["Cancelled_Flights"]
      .sum()
      .sort_values(ascending=False)
)

cancel
```

```python
plt.figure(figsize=(12,6))

cancel.plot(kind="bar", color="orange")

plt.title("Cancelled Flights by Airline")

plt.xticks(rotation=90)

plt.show()
```

Delay Severity Distribution

```python
severity = df["Delay_Severity"].value_counts()

severity
```

```python
severity.plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6)
)

plt.ylabel("")
plt.title("Delay Severity Distribution")

plt.show()
```

Compare 2019 vs 2020

```python
comparison = (
    df.groupby("Year")[[
        "Total_Flights",
        "Delayed_Flights",
        "Cancelled_Flights"
    ]]
    .sum()
)

comparison
```

```python
comparison.plot(
    kind="bar",
    figsize=(10,6)
)

plt.title("2019 vs 2020 Flight Operations")

plt.xticks(rotation=0)

plt.show()
```
